Cleaning Size Column

```python
df['Size'].unique()
```

```python
# procedure for converting the column "Size" to float
# there are sizes counted in mb, kb, in numbers without measurement unit and with "varies with device"

# removing the "m" which is the mb for the size
df['Size']= df["Size"].str.replace("M", "")
df['Size']= df['Size'].str.replace("Varies with device","-1")

#Then we replace k and change the unit to Mb
df['Size']=df['Size'].apply(lambda x: str(round((float(x.rstrip('k'))/1024),1) )if x[-1]=='k' else x)  # conversion of 1 mb into kb by diving it by 1024 because 1 mb = 1024 kb 
df['Size'] =df['Size'].astype(float)

# Checking unique values
df['Size'].unique()
```

dropna(subset="column name", inplace= True) method 

The dropna() function is used to drop null values.
The subset parameter in the dropna method allows you to specify a subset of columns in which to drop the null values. For example, if you have a DataFrame with columns A, B, and C, you can use subset=['B', 'C'] to drop the null values only from the B and C columns.
On the other hand, when you use dropna(inplace=True), it drops the null values from the entire DataFrame or Series inplace, which means that it modifies the original object itself rather than returning a new object

Since there is only 8 null values so we will drop these null values by using dropna() function.

```python
# Dropping null values in 'Current Ver' Column
df.dropna(subset=['Current Ver'],inplace=True)
df
```

## Filling missing values in the Categorical columns.

1. Here, we realized that there are 987 rows having null values under column 'Translated_Review'. Hence, we replace the null values with mode of Translated_Review by using fillna method.

```python
df1['Translated_Review']=df1['Translated_Review'].fillna(df1['Translated_Review'].mode()[0])
df1['Translated_Review']
```

```python
# Checking if there is any null value in the column 'Translated_Review' or not
df1['Sentiment'].isna().any()
```

## Filling Missing Values in the Numerical Column

1. There are 982 null values in the column 'Sentiment_Polarity'. Hence we replace the null values with median of overall 'Sentiment_Polarity' values.

2. There are 982 null values in the column 'Sentiment_Subjectivity'. Hence we replace the null values with median of overall 'Sentiment_Subjectivity' values.

```python
# Defining a function
def replace_with_median(series):
    """
    Given a series, replace the rows with null values 
    with median values
    """
    return series.fillna(series.median())  # returns the median value

# Using transform method to execute a function for each value of the Dataframe
df1[['Sentiment_Polarity','Sentiment_Subjectivity']] = df1[['Sentiment_Polarity','Sentiment_Subjectivity']].transform(replace_with_median)
# Round the values upto 2 decimal places
df1[['Sentiment_Polarity','Sentiment_Subjectivity']].astype(float).round(2)
```
